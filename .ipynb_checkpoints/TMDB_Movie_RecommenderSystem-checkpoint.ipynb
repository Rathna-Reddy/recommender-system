```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies.head(1)['genres'].values
```

```python
credits.head(1)['cast'].values
```

```python
print('movies shape: ',movies.shape)
print('credits shape: ',credits.shape)
```

```python
#merge movies and credits into a single file using movie_id
movies = movies.merge(credits, on=credits['movie_id'])
print(movies.shape)
```

```python
movies.head(5)
```

```python
#genres
#id
#keywords
#overview
#title_x
#cast
#crew
movies = movies[['id', 'title_x', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies = movies.rename(columns={'id' : 'movie_id', 'title_x' : 'title'})
movies.head(5)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
print(movies.shape)
```

```python
#check if there any duplicated rows
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
#in the genres column cell data, we will extract genre name values as a list
#here we are using ast.literal_eval to convert String object into list of dictionaries.
extract_genres = lambda genres: [genre['name'] for genre in ast.literal_eval(genres)]

#Apply lambda function to each cell in the 'genre' column
movies['genres'] = movies['genres'].apply(extract_genres)

movies.head(5)
```

```python
movies.iloc[0].keywords
```

```python
#in the same way we will extract for keywords
extract_keywords = lambda keywords: [keyword['name'] for keyword in ast.literal_eval(keywords)]

movies['keywords'] = movies['keywords'].apply(extract_keywords)
movies.head(5)
```

```python
movies.iloc[0].cast
```

```python
#For the cast we are interetested in only first 3 names
extract_cast = lambda cast: [ca['name'] for ca in ast.literal_eval(cast)[:3]]
movies['cast'] = movies['cast'].apply(extract_cast)

movies.head(5)
```

```python
movies.iloc[0].crew
```

```python
#in the crew column data, we will get name where job='Director'
extract_director = lambda crew: [crew_m['name'] for crew_m in ast.literal_eval(crew) if crew_m['job'] == 'Director']

movies['crew'] = movies['crew'].apply(extract_director)

movies.head(5)
```

```python
movies['overview'][0]
```

```python
#we will convert this string to list to be able to easily concatenate later.
movies['overview'] = movies['overview'].apply(lambda x: x.split(' '))
movies.head()
```

```python
#we gonna remove ' ' and replace it with '' to make them unique
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
movies.head(5)
```

```python
#let's concatenate overview, genres, keywords, cast and crew columns data into a single column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(5)
```

```python
#now lets create a new data frame with just movie_id, title and tags columns
df_movies = movies[['movie_id', 'title', 'tags']]
df_movies.head(5)
```

```python
#now lets convert list of tags to string inorder to perform other pre-processing steps like stemming, removal of stopwords etc.
df_movies['tags'] = df_movies['tags'].apply(lambda x: ' '.join(x))
df_movies.head(5)
```

```python
df_movies['tags'][0]
```

```python
#convert all the words to lower case
df_movies['tags'] = df_movies['tags'].apply(lambda x: x.lower())
df_movies['tags'][0]
```

```python
df_movies.head(5)
```

```python
#here we will do stemming to get the root word
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def perform_stemming(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(perform_stemming('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron'))
```

```python
df_movies['tags'] = df_movies['tags'].apply(perform_stemming)
```

```python
#perform vectorization to tags column using bag of words.
#we also remove the stop words.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

bow_tags = cv.fit_transform(df_movies['tags']).toarray()
print(bow_tags.shape)
```

```python
#get the last 10 words in the vector
cv.get_feature_names_out()[-10:-1]
```

```python
#to get the distance between vectors we gonna use cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
movies_similarity = cosine_similarity(bow_tags)

print(movies_similarity.shape)
```

```python
print(movies_similarity)
```

```python
def recommend_similar_movies(movie):
    movie_index = df_movies[df_movies['title'] == movie].index[0]
    similarity_dist_for_a_movie = movies_similarity[movie_index]
    sorted_movie_list = sorted(list(enumerate(similarity_dist_for_a_movie)), reverse=True, key=(lambda x: x[1]))[1:6]
    
    for i in sorted_movie_list:
        print(df_movies.iloc[i[0]].title)
    
```

```python
recommend_similar_movies('Batman Begins')
```
